## Cuvinte cheie: pass, continue, break

```python
pass -> expresie nula; 
continue -> continua catre urmatoarea iteratie a unei bucle
break -> forteaza iesire din bucla
```

```python
for char in "Propozitie":
    if char == 'o':
        pass
    print(char)
```

```python
for char in "Propozitie":
    if char == 'o':
        continue
    print(char)
```

```python
for char in "Propozitie":
    if char == 'o':
        break
    print(char)
```

```python
lista_credit = []
lista_debit = []

for operatiune in lista_operatiuni:
    if operatiune > 0:
        lista_debit.append(operatiune)
    elif operatiune < 0:
        lista_credit.append(operatiune)
    else:
        pass

print(f"Lista operatiunilor de debit este: {lista_debit}, iar lista operatiunilor de credit este: {lista_credit}")
```

```python
lista_credit = []
lista_debit = []

for operatiune in lista_operatiuni:
    if operatiune == 0:
        break
        print("A fost intalnita o operatiune nula")
    elif operatiune > 0:
        lista_debit.append(operatiune)
    else:
        lista_credit.append(operatiune)
print(f"Lista operatiunilor de debit este: {lista_debit}, iar lista operatiunilor de credit este: {lista_credit}")
```

### Tema Generare date aleatorii.
Sa se genereze 100k tranzactii, cu urmatoarele atribute: id_tranzactie, produs, cantitate, pret.
- __id_tranzactie__: secventa numerica, de la 1 la 100k, incrementata cu 1
- __produs__: format -> grupa_produs + id_produs + tara_produs;
    - __grupa_produs__: N01, N02, N03, N04, N05
    - __id_produs__: secventa numerica de la 1001 la 1200, incrementata cu 1
    - __tara_produs__: DE, FR, UK, RO, CH, NL, BE, IT, ES, PR
    - __indiciu__: pentru fiecare element (grupa_produs, id_produs, tara_produs) din format va trebui generata o lista separata de valori, care pot fi unite apoi, cu ajutorul metodei .join(), asociata string-urilor, dupa ce listele vor fi unite.
- __cantitate__: secventa numerica, de la 1 la 5, incrementata cu 1
- __pret__: secventa numerica, de la 12.49 la 37.00 incrementata cu 0.5.

Pentru usurinta generarii listelor, pot fi folosite frecvente uniforme, la nivel de lista. Ex. pt atributul __cantitate__, care ia valori de la 1 la 5, fiecare din valori poate aparea in lista de exact 20k ori. Lungimea fiecarei liste a fost conceputa incat sa fie divizor al lungimii intregii liste: 100k.<br> 
__Nota__: Valorile din cele 4 liste de atribute vor trebui amestecate, pentru aleatorizare. Apoi listele vor trebui unite sub o singura lista de inregistrari.<br>
Sa se determine suma vanzarilor realizate, pentru Romania.<br>
__Bonus__: Sa se defineasca o functie, care sa genereze fiecare din listele necesare, pe baza unei liste de valori de intrare.
