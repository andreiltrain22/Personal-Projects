```python
select top 1
    cust.customer_state, 
    count(distinct orders.order_id) as no_orders, 
    count(distinct cust.customer_unique_id) as no_customers, 
    sum(price) as sales
from 
    customers_df as cust
        innner join orders_df as orders on orders.customer_id = cust.customer_id
        inner join order_items as items on items.order_id = orders.order_id
group by 
    cust.customer_state
order by 
    no_orders desc
```

```python
def sort_by_column(column_name, sort_direction):
    return (
    customers_df[['customer_state', 'customer_id', 'customer_unique_id']]
        .merge(right = orders_df[['customer_id', 'order_id']], how = 'inner', on = 'customer_id')
        .merge(right = order_items_df[['order_id', 'price']], how = 'inner', on = 'order_id')
        .groupby(by = 'customer_state')
        .agg(
             {'order_id':'nunique',
              'customer_unique_id':'nunique',
              'price':'sum'}
            )
        .rename(columns = {'order_id':'no_orders',
                           'customer_unique_id':'no_customers',
                           'price':'sales'}
               )
        .sort_values(by = column_name, ascending = sort_direction)
        .iloc[:1] #.head(1)
    )
```

```python
sort_by_column('no_orders', False)
```

```python
sort_by_column('no_customers', False)
```

```python
sort_by_column('sales', False)
```

```python
sort_by_column('no_orders', True)
```

```python
sort_by_column('no_customers', True)
```

```python
sort_by_column('sales', True)
```
