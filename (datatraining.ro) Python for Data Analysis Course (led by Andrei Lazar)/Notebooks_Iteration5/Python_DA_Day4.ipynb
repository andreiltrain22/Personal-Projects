### Metoda .apply()

```python
def last_four_digits(num):
    return int(str(num)[-4:])
```

```python
tips_df['last_four_digits'] = tips_df['CC Number'].apply(func = last_four_digits)
tips_df
```

```python
def price_category(price):
    if price < 10:
        return 'ieftin'
    elif price >= 10 and price < 30:
        return 'mediu'
    else:
        return 'scump'
```

```python
tips_df['price_category'] = tips_df['total_bill'].apply(func = price_category)
tips_df[tips_df['price_category'] != 'mediu'][['total_bill', 'price_category']]
```

```python
tips_df['price_category'] = tips_df['total_bill'].apply(func = lambda price: 'ieftin' if price < 10 
                                                                                      else (
                                                                                        'mediu' if price >= 10 and price <= 30 
                                                                                                else 'scump')
                                                       )
```

```python
tips_df[['total_bill', 'price_category']].sample(frac = 0.1)
```

```python
tips_df['tip_quality'] = tips_df[['total_bill', 'tip']].apply(
                                            func = lambda tips_df: 'Generous' if tips_df['tip']/tips_df['total_bill'] > 0.25
                                                                              else 'Not Generous',
                                             axis = 1
                                                )
```

```python
tips_df[['total_bill', 'tip', 'tip_quality']].sample(n=15)
```

Sa se calculeze top 5 categorii de produse comandate, dupa numarul de comenzi.

```python
#prod_ctg_translation_df -> products_df -> order_items_df -> count(distinct order_id)
```

```python
select top 5 ctg.product_category_name_english, count(distinct items.order_id) as no_orders
from products_df as prod
inner join prod_ctg_translation_df as ctg on ctg.product_category_name = prod.product_category_name
inner join order_items_df as items on items.product_id = prod.product_id
group by ctg.product_category_name_english
order by no_orders desc
```
