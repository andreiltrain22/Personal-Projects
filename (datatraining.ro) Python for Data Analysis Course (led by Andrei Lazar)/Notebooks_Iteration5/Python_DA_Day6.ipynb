### Import/export baze de date relationale

```python
import sqlite3 as sq3
```

```python
connection = sq3.connect(database = 'olist_orders.db')
```

```python
connection.execute(" select name from sqlite_master where type = 'table' ").fetchall()
```

```python
import os
```

```python
os.listdir(path)
```

```python
for file_name in os.listdir(path):
    if 'dataset.csv' in file_name or 'translation' in file_name:
        (pd.read_csv(path+file_name) #-->returneaza dataframe
            .to_sql(name = file_name.replace('olist_', '')
                                    .replace('.csv', '')
                                    .replace('_dataset', ''), 
                    con = connection, 
                    index = False)
        )
```

```python
pd.read_sql(sql = "select * from sqlite_master where type = 'table' ", con = connection)
#alternativ: connection.execute(" select name from sqlite_master where type = 'table' ").fetchall()
```

```python
connection.execute(" select name from sqlite_master where type = 'table' ").fetchall()
```

```python
pd.read_sql(sql = """
                     select count(*)
                     from orders
                     where order_status = 'delivered'
                    """, con = connection)
```

```python
pd.read_sql(sql = """
                    select order_status, count(*) as cnt
                    from orders
                    group by order_status
                    order by cnt desc
                    """, con = connection)
```

```python
for name in pd.read_sql(sql = "select name from sqlite_master where type = 'table' ", con = connection)['name']:
    print(name)
```

```python
for table in pd.read_sql(sql = "select name from sqlite_master where type = 'table' ", con = connection)['name']:
    print(pd.read_sql(f"select '{table}' as table_name, count(*) as cnt from {table}", con = connection), '\n')
```
