```
# Use Bronze Store DB to retrieve data from station table
spark.sql("USE DATABASE bronze_store")
```

```
# Load data from station table to dataframe
station_df = (
spark
    .read
    .table(tableName = "station")
)

display(station_df)
```

```
from pyspark.sql.functions import col
from pyspark.sql.types import FloatType

# Convert latitude and longitude columns to float type
station_df = (
    station_df
        .select(
            col("station_id"),
            col("station_name"),
            col("latitude").cast(FloatType()),
            col("longitude").cast(FloatType())
            )
)

station_df.printSchema()
```

```
# Use Gold Store to load dimension tables and fact tables that are part of the designed star schema
spark.sql("USE DATABASE Gold_Store")
```

```
# Load dataframe content to dimension table inside the Gold Store
(
station_df
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("dim_station")
)
```
